## Data Exploration using Seaborn

## Import the iris Dataset

```python
import pandas as pd
import seaborn as sns
from sklearn.datasets import load_iris
iris=load_iris()
data=pd.DataFrame(iris.data,columns=iris.feature_names)
label=pd.DataFrame(list(map(lambda x : iris.target_names[x],iris.target)),columns=['Species'])
iris=pd.concat([data,label],axis=1)
iris.head()
```

```python
data.head()
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

## 1. Displot -  Use the displot() to see the distribution of the SepalLengthCm,  SepalWidthCm, PetalLengthCm, PetalWidthCm features. Plot them as single image

Declared one plot with 4 subplots inside and displayed one parameter in each subplot.

```python
fig,axes=plt.subplots(2,2,figsize=(20,10))
sns.distplot(iris['sepal length (cm)'],ax=axes[0,0])
sns.distplot(iris['sepal width (cm)'],ax=axes[0,1])
sns.distplot(iris['petal length (cm)'],color='g',ax=axes[1,0])
sns.distplot(iris['petal width (cm)'],color='g',ax=axes[1,1])
plt.show()
```

## 2. Boxplot - Do box plot of all the features without 'Species'

Declared one plot of size (20,10) and plotted box plot using boxplot keyword


```python
fig = plt.figure(figsize=(20,10))
sns.boxplot(data=[iris['sepal length (cm)'],iris['sepal width (cm)'],iris['petal length (cm)'],iris['petal width (cm)']],palette='rainbow')
plt.title("Box Plot")
plt.xlabel("Sepal length (cm), Sepal width (cm), Petal length (cm), Petal width (cm)")
```

## 3. CountPlot - Do a countplot for the feature 'Species'

Plotted the countplot for Species variable.

```python
sns.countplot(x='Species',data=iris)
```

## 4. PairPlot - Do a pairplot on the features 'SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm', 'Species

Plotted pairplot and declared species column as hue and other columns as vars

```python
sns.pairplot(iris, hue='Species', size=2.5, vars=["sepal length (cm)","sepal width (cm)","petal length (cm)","petal width (cm)"])
```

## 5. lmplot - Do an implot on the following SepalLengthCm, PetalLengthCm. Using hue, display the different species in different colors

Plotted lmplot with hue as species and Sepal length and Petal Length as x and y variables

```python
sns.lmplot("sepal length (cm)","petal length (cm)",hue="Species",data=iris)
```

## 6. BarPlot - Do a Bar plot of 'Species' vs 'SepalLegthCm'

Plotted barplot for Species and Sepal Length varaibles

```python
sns.barplot("Species","sepal length (cm)",data=iris)
```

## 7. HeatMap - Using heatmap, plot the corration matrix.

```python
sns.heatmap(iris.corr(),cmap='coolwarm')
plt.title("iris.corr()")
```
